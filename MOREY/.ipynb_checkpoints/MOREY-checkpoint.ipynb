```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

raptors = ["#00B2A9","#BA0C2F", "#753BBD"]
rockets = ["#000000", "#C4CED4","#CE1141"]
color = ["#753BBD","#FF5722", "#009688"]

sns.color_palette(raptors)
%matplotlib inline
```

```python
df = pd.read_csv('nba17-18.csv')
df.rename(columns={'3:00 PM':'3PM'}, inplace=True)
df = df[df.FGA >= 150]
print(df.columns.values)
df2 = df[['PLAYER','TEAM','FGA','3PM','RAFGM','FTM']]

fa = pd.read_csv('FA18.csv')
fa.rename(columns={'Team':'TEAM', 'Type':'TypeFA'}, inplace=True)
def reverseName(name):
    arr = name.split(', ')
    return " ".join(list(reversed(arr)))
fa['PLAYER'] = fa['Player'].apply(reverseName)
fa.drop('Player', axis=1, inplace=True)
df2 = df2.merge(fa, on=['PLAYER', 'TEAM'], how='left')
df2.head(30)
```

```python
df2['MOREY'] = (3*df2['3PM'] + 2*df2.RAFGM + df2.FTM)/df2.FGA
df2['3P'] = 3*df2['3PM']/df2.FGA
df2['RA'] = 2*df2.RAFGM/df2.FGA
df2['FT'] = df2.FTM/df2.FGA
dfSorted = df2.sort_values('MOREY', ascending=False).reset_index(drop=True)
df3 = dfSorted.iloc[:20,:]
df3
```

```python
sns.set_palette(sonics)
# sns.set()
bar = df3[['PLAYER','3P','FT','RA']]
bar.set_index('PLAYER', inplace=True)
ax = bar.plot.bar(stacked=True, width=.8,figsize=(10,10))
ax.tick_params(direction='out', length=1, width=2, labelsize='large')
ax.set_title('Top MOREY Scores', fontsize=20)
ax.legend( prop={'size': 15})
plt.savefig('top-morey.png',format='png')
```

```python
hou = dfSorted[dfSorted.TEAM == 'HOU']

sns.set_style('whitegrid')
bar = hou[['PLAYER','3P','FT','RA']]
bar.set_index('PLAYER', inplace=True)
nohou = dfSorted[dfSorted.TEAM != 'HOU']
topfa = nohou.dropna()
topfa = topfa.iloc[:30,:]
bar3 = topfa[['PLAYER','3P','FT','RA']]
bar3.set_index('PLAYER', inplace=True)

f, (ax1, ax2) = plt.subplots(1, 2,  sharey=True, gridspec_kw = {'width_ratios':[1, 2]},)
f.set_size_inches(18,18)
sns.set_palette(rockets)
sns.set_style('whitegrid')
ax1 = bar.plot.bar(stacked=True, width=.8, ax=ax1)
ax1.tick_params(direction='out', length=1, width=2, labelsize='xx-large')
ax1.set_title('HOU MOREY Scores', fontsize=20)
ax1.legend( prop={'size': 20})

sns.set_palette(color)
ax2 = bar3.plot.bar(stacked=True, width=.8, ax=ax2)
ax2.tick_params(direction='out', length=1, width=2, labelsize='xx-large')
ax2.set_title('Top Free Agent MOREY Scores', fontsize=20)
ax2.legend( prop={'size': 20})

plt.tight_layout()
plt.savefig('fa-morey.png',format='png')
```

```python
nohou = dfSorted[dfSorted.TEAM != 'HOU']
topfa = nohou.dropna()
sns.set_palette(sonics)
topfa = topfa.iloc[:30,:]
bar3 = topfa[['PLAYER','3P','FT','RA']]
bar3.set_index('PLAYER', inplace=True)
ax = bar3.plot.bar(stacked=True, width=.8,figsize=(10,10))
ax.tick_params(direction='out', length=1, width=2, labelsize='large')
ax.set_title('Top Free Agent MOREY Scores')
```

```python
dfg = dfSorted.groupby('TEAM', as_index=False).mean()
dfg.sort_values('MOREY', inplace=True, ascending=False)
dfg.reset_index(inplace=True, drop=True)
dfg[['TEAM','MOREY']]
```

```python
plt.figure(figsize=(10,10))
bar2 = dfg[['TEAM','3P','FT','RA']]
bar2.set_index('TEAM', inplace=True)
ax = bar2.plot.bar(stacked=True, width=.8,figsize=(12,12))
ax.tick_params(direction='out', length=1, width=2, labelsize='xx-large')
ax.set_title('Team MOREY Scores', fontsize=20)
ax.legend( prop={'size': 20})
plt.savefig('team-morey.png',format='png')
```

$$\huge MOREY\ Score = \frac{3\cdot 3PM + 2\cdot RAM + FTM}{FGA}$$
