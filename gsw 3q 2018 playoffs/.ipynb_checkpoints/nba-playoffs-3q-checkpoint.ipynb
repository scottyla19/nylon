```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from pandas.tools.plotting import table

flatui = ["#006BB6", "#FDB927", "#26282A"]
sns.set_palette(flatui)
```

```python
df = pd.read_csv('GSW by Quarter.csv')
df.head()
```

```python
gsw = df[df['TEAM'] == "Golden State Warriors"]
gsw.drop('TEAM', axis=1,inplace=True)
opp = df[df['TEAM'] != "Golden State Warriors"]
byQuarter = opp.groupby('Q', as_index=False).mean()
# byQuarter.dtypes
```

```python
bar_labels = byQuarter.drop(['FT%','Q'],axis=1).columns.values
def getGSWQuarter(q):
    x1 = gsw[gsw['Q']== q]
    x1.drop(['FT%','Q'],inplace=True,axis=1)
    return x1.iloc[0,:]
    
def getOPPQuarter(q):
    x2 = byQuarter[byQuarter['Q']== q]
    x2.drop(['FT%','Q'],inplace=True,axis=1)
    x2=x2.round(1)
    return x2.iloc[0,:]
```

```python
def makePlot(x1, x2,q):
    fig = plt.figure(figsize=(8,6))
    # Set the y position
    y_pos = np.arange(len(x1))
    y_pos = [x for x in y_pos]
    plt.yticks(y_pos, bar_labels, fontsize=10)

    # Create a horizontal bar in the position y_pos
    plt.barh(y_pos, 
             # using x1 data
             x1, 
             # that is centered
             align='center', 
             # with alpha 0.4
             alpha=0.4,
             color='#006BB6')

    # Create a horizontal bar in the position y_pos
    plt.barh(y_pos, 
             # using NEGATIVE x2 data
             -x2,
             # that is centered
             align='center', 
             # with alpha 0.4
             alpha=0.4,
             color='#CE1141')

    # annotation and labels
    plt.xlabel('{}Q'.format(q))
    plt.xticks([0])
    for i, v in enumerate(x1):
        plt.text(v, i , str(v),verticalalignment='center' , color='#006BB6', fontweight='bold')
        
    for i, v in enumerate(-x2):
        if i < len(x2) - 1:
            plt.text(v - 4, i , str(abs(v)),verticalalignment='center' ,horizontalalignment='left', color='#CE1141', fontweight='bold')
        
    t = plt.title('GSW vs Opponents in NBA Playoffs 2018')
    plt.ylim([-1,len(x1)+0.1])
    plt.xlim([-max(x2)-6, max(x1)+6])
    plt.legend(['GSW','Opponents'])
    plt.grid(axis='x')
    plt.savefig('{}Q-comparison.png'.format(q))
    
    
```

```python
# fig = plt.figure(figsize=(8,6))

# plt.subplot(221)
makePlot(getGSWQuarter(1), getOPPQuarter(1), 1)
# plt.subplot(222)
makePlot(getGSWQuarter(2), getOPPQuarter(2), 2)
# plt.subplot(223)
makePlot(getGSWQuarter(3), getOPPQuarter(3), 3)
# plt.subplot(224)
makePlot(getGSWQuarter(4), getOPPQuarter(4), 4)

# plt.savefig('2018-GSW-3Q-dominance.png')
```

```python
df2 = pd.read_csv('GSW-Players.csv')
df2.head()
```

```python
big3 = df2[df2['PLAYER'].isin(["Kevin Durant",'Klay Thompson','Stephen Curry'])]
big3['TS%'] = big3['PTS']/(2*(big3['FGA'] + .44*big3['FTA']))
big3['eFG%']= (big3.FGM + (.5*big3['3PM']))/big3.FGA  
big3Q = big3.groupby('Q', as_index=False )
big3qShoot = big3[['Q','PLAYER','PTS','FGA','FG%','3PA','3P%','FTA','FT%','TS%','eFG%']]
big3qShoot.head()

```

```python
flatui = ["#006BB6", "#FDB927", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
sns.set_palette(flatui)
g = sns.factorplot(x="Q", y="PTS", hue="PLAYER", data=big3qShoot,
                   size=6, kind="bar")
# g.despine(left=True)
g.set_ylabels("PTS")
g.ax.set_title("PTS by Quarter")

g.savefig("pts.png")

g = sns.factorplot(x="Q", y="TS%", hue="PLAYER", data=big3qShoot,
                   size=6, kind="bar")
g.set_ylabels("TS%")
g.ax.set_title("TS% by Quarter")
g.savefig("ts.png")
```
