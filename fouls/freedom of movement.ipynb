```python
teamsPrev = df[df.SEASON != '2018-19']
teamsCurrent = df[df.SEASON == '2018-19']
teamsAvg = teamsPrev.groupby("TEAM", as_index=False).mean()
fullTmNames = teamsCurrent.TEAM.unique()
tmColorDict = dict(zip(fullTmNames, fillColors))
tmAbbvDict = dict(zip(fullTmNames, teamAbbvs))
# teamsCurrent.drop("SEASON",axis=1, inplace=True)
```

```python

teams1718 = df[df.SEASON == '2017-18']

# teams1718.drop("SEASON",axis=1, inplace=True)
```

```python
teams2 = teams1718.merge(teamsCurrent,on='TEAM', suffixes=('_Avg','_Current'))

teams2['FTM_Diff'] = teams2.FTM_Current - teams2.FTM_Avg 
teams2['FTA_Diff'] = teams2.FTA_Current - teams2.FTA_Avg 
teams2['PF_Diff'] = teams2.PF_Current - teams2.PF_Avg 
teams2['PFD_Diff'] = teams2.PFD_Current - teams2.PFD_Avg 
# teams['%PTS FT_Diff'] = teams.FTM_Current - teams.FTM_Avg 
teams2['PACE_Diff'] = teams2.PACE_Current - teams2.PACE_Avg 
teams2['PFD/POSS_Diff'] = teams2['PFD/POSS_Current'] - teams2['PFD/POSS_Avg'] 
teams2['FTA/POSS_Diff'] = teams2['FTA/POSS_Current'] - teams2['FTA/POSS_Avg'] 
```

```python
teams2.sort_values('FTA/POSS_Diff',ascending=False)

```

```python

last2years = df[(df.SEASON == '2017-18') | (df.SEASON == '2018-19')]
last2diffs = last2years.merge(teams2[['TEAM',"FTM_Diff","FTA_Diff", 'PF_Diff',"PFD_Diff","PACE_Diff","PFD/POSS_Diff",'FTA/POSS_Diff']],on='TEAM')
last2diffs['FTA_Change'] = np.where(last2diffs['FTA_Diff']>=0, 'Increase', 'Decrease')
last2diffs['FTA_Change'] = np.where(last2diffs['FTA_Diff']>=3, 'Large Increase', last2diffs['FTA_Change'])
last2diffs['PFD/POSS_Change'] = np.where(last2diffs['PFD/POSS_Diff']>=0, 'Increase', 'Decrease')
last2diffs['PFD/POSS_Change'] = np.where(last2diffs['PFD/POSS_Diff']>=2, 'Large Increase', last2diffs['PFD/POSS_Change'])

last2diffs['FTA/POSS_Change'] = np.where(last2diffs['FTA/POSS_Diff']>=0, 'Increase', 'Decrease')
last2diffs['FTA/POSS_Change'] = np.where(last2diffs['FTA/POSS_Diff']>=3, 'Large Increase', last2diffs['FTA/POSS_Change'])
# last2diffs.rename(index=str, columns={"PFD/POSS": "PFD_Per_100_POSS"}, inplace=True)
```

```python
def addTmAbbv(ax, col, colVal):
    colChange = col+'_Change'
    df1 = last2diffs[(last2diffs[colChange] == colVal) ].reset_index()
    for line in range(0,df1.shape[0]):
         ax.text(df1.SEASON[line], df1[col][line] , tmAbbvDict[df1.TEAM[line]], horizontalalignment='left', size='medium', color=tmColorDict[df1.TEAM[line]], weight='semibold')

            
fig, axs = plt.subplots(nrows=2, ncols=3, sharey='row')
ax1 = sns.lineplot(x='SEASON', y='FTA/POSS', data=last2diffs[last2diffs['FTA/POSS_Change'] == "Large Increase"], hue="TEAM", ax=axs[0,0], legend=False, palette=tmColorDict)
ax2 =sns.lineplot(x='SEASON', y='FTA/POSS', data=last2diffs[last2diffs['FTA/POSS_Change']  == "Increase"], hue="TEAM", ax=axs[0,1], legend=False, palette=tmColorDict)
ax3 =sns.lineplot(x='SEASON', y='FTA/POSS', data=last2diffs[last2diffs['FTA/POSS_Change']  == "Decrease"], hue="TEAM", ax=axs[0,2], legend=False, palette=tmColorDict)
ax4 =sns.lineplot(x='SEASON', y='PFD/POSS', data=last2diffs[last2diffs['PFD/POSS_Change'] == "Large Increase"], hue="TEAM", ax=axs[1,0], legend=False, palette=tmColorDict)
ax5 =sns.lineplot(x='SEASON', y='PFD/POSS', data=last2diffs[last2diffs['PFD/POSS_Change'] == "Increase"], hue="TEAM", ax=axs[1,1], legend=False, palette=tmColorDict)
ax6 =sns.lineplot(x='SEASON', y='PFD/POSS', data=last2diffs[last2diffs['PFD/POSS_Change']  == "Decrease"], hue="TEAM", ax=axs[1,2], legend=False, palette=tmColorDict)

addTmAbbv(ax1,'FTA/POSS','Large Increase')
addTmAbbv(ax2,'FTA/POSS','Increase')
addTmAbbv(ax3,'FTA/POSS','Decrease')
addTmAbbv(ax4,'PFD/POSS','Large Increase')
addTmAbbv(ax5,'PFD/POSS','Increase')
addTmAbbv(ax6,'PFD/POSS','Decrease')
    

fig.set_figheight(15)
fig.set_figwidth(15)
plt.savefig('Team-FTA-PFD-change-2018-19.png',bbox_inches='tight')
```
